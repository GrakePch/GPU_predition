````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
def analysis(data):

    df = pd.DataFrame(data, columns=['Size', 'Predicted Time', 'Actual Time'])

    plt.figure(figsize=(8, 5))
    plt.plot(df['Size'], df['Predicted Time'], label='Predicted Time', marker='o')
    plt.plot(df['Size'], df['Actual Time'], label='Actual Time', marker='x')
    plt.xlabel('Threads')
    plt.ylabel('Time')
    plt.title('Predicted Time vs Actual Time')
    plt.legend()
    plt.grid(True)
    plt.ylim(0, )
    plt.show()

    # percentage error
    df['Accuracy'] = 100 - abs((df['Predicted Time'] - df['Actual Time']) / df['Actual Time']) * 100

    plt.figure(figsize=(8, 5))
    plt.plot(df['Size'], df['Accuracy'], label='Percentage Error', marker='o', color='r')
    plt.xlabel('Threads')
    plt.ylabel('Percentage Error (%)')
    plt.title('Percentage Error Between Predicted and Actual Time')
    plt.legend()
    plt.grid(True)
    # plt.ylim(0, )
    plt.show()
````

# vecaddf: vecf + vecf

````python
# [element-in-vector, predicted-time, actual-time]
data = [
    [32000000, 0.000218,   0.000635],
    [64000000, 0.000432,   0.001264],
    [128000000, 0.000861,  0.002521],
    [256000000, 0.001718,  0.005046],
]
analysis(data)
````
